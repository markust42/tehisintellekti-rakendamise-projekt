# Andmetega tutvumine

See töövihik tutvustab toorandmestiku põhiomadusi.

## 1. Andmete laadimine

```python
import pandas as pd
import numpy as np

# Laadime toorandmed
df = pd.read_csv('toorandmed.csv')

print("Andmed edukalt laaditud!")
```

## 2. Tunnuste ja andmepunktide arv toorandmestikus

```python
# Arvutame andmepunktide (ridade) ja tunnuste (veergude) arvu
andmepunkte = df.shape[0]
tunnuseid = df.shape[1]

print("="*60)
print("TOORANDMESTIKU SUURUS")
print("="*60)
print(f"Andmepunktide arv (read):  {andmepunkte:,}")
print(f"Tunnuste arv (veerud):     {tunnuseid}")
print("="*60)
```

## 3. Ühe andmerea (ühe aine) põhjalik läbivaatus

Vaatame põhjalikult läbi ühe aine informatsiooni ja teeme tähelepanekuid tunnuste vormindamise kohta.

```python
# Valime juhusliku andmerea
import random
juhuslik_index = random.randint(0, len(df) - 1)
aine = df.iloc[juhuslik_index]

print("="*80)
print(f"ÜKSIKASJALIK ÜLEVAADE ÜHEST AINEST (rida {juhuslik_index + 1})")
print("="*80)
print()

# Kuvame kõik tunnused ja nende väärtused
for tunnus, vaartus in aine.items():
    print(f"{tunnus}:")
    print(f"  {vaartus}")
    print()
```

## 4. Tunnuste vormindamise info

Vaatame tunnuste omadusi, et saaksid ise järeldusi teha.

```python
print("="*80)
print(f"VALITUD AINE TUNNUSTE VORMINDAMISE INFO (rida {juhuslik_index + 1})")
print("="*80)
print()

# 1. Täidetud vs tühjad tunnused
print("1. TÄIDETUD JA TÜHJAD TUNNUSED:")
taiedetud = aine.notna().sum()
tuhjad = aine.isna().sum()
print(f"   Täidetud tunnuseid: {taiedetud}/{len(aine)}")
print(f"   Tühje tunnuseid: {tuhjad}/{len(aine)}")
print()

# 2. Andmetüübid selles reas
print("2. ANDMETÜÜBID:")
for dtype in aine.dtype.categories if hasattr(aine.dtype, 'categories') else [aine.dtype]:
    count = sum(1 for val in aine if pd.notna(val) and type(val).__name__ == dtype.name)
print(f"   {aine.dtype}")
print()

# 3. Mitmekeelsed tunnused selles reas
print("3. MITMEKEELSED TUNNUSED (täidetud):")
keel_tunnused_taiedetud = [(col, aine[col]) for col in aine.index if ('__et' in col or '__en' in col or '__ru' in col) and pd.notna(aine[col])]
print(f"   Täidetud mitmekeelseid tunnuseid: {len(keel_tunnused_taiedetud)}")
if len(keel_tunnused_taiedetud) > 0:
    print(f"   Näiteid:")
    for col, val in keel_tunnused_taiedetud[:5]:
        preview = str(val)[:60] + "..." if len(str(val)) > 60 else str(val)
        print(f"      {col}: {preview}")
print()

# 4. JSON-vormingus tunnused selles reas
print("4. JSON-VORMINGUS TUNNUSED (täidetud):")
json_tunnused_taiedetud = []
for col in aine.index:
    if pd.notna(aine[col]):
        val = str(aine[col])
        if val.startswith('[') or val.startswith('{'):
            json_tunnused_taiedetud.append(col)
print(f"   Täidetud JSON tunnuseid: {len(json_tunnused_taiedetud)}")
if len(json_tunnused_taiedetud) > 0:
    print(f"   Näiteid: {json_tunnused_taiedetud[:5]}")
print()

# 5. Tunnuste eesliited (täidetud)
print("5. TÄIDETUD TUNNUSTE JAOTUS EESLIIDETE KAUPA:")
prefixes_taiedetud = {}
for col in aine.index:
    if pd.notna(aine[col]):
        prefix = col.split('__')[0]
        if prefix not in prefixes_taiedetud:
            prefixes_taiedetud[prefix] = 0
        prefixes_taiedetud[prefix] += 1

for prefix, count in sorted(prefixes_taiedetud.items(), key=lambda x: x[1], reverse=True)[:10]:
    print(f"   {prefix}: {count}")
print("="*80)
```
